Program the maximum subarray problem in python.

Seek to plot the actual time spent and compare with the theoretical complexity we learn in the class.  Attempt to compare many sizes of n.   Are the actual time close with the theoretical complexity?

```python
import numpy as np
import matplotlib.pyplot as plt
import time

def find_max_crossing_subarray(A,low,mid,high):
    left_sum = -np.inf
    s = 0
    for i in range(mid,low-1,-1):
        s += A[i]
        if(s > left_sum):
            left_sum = s
            max_left = i
    right_sum = -np.inf
    s = 0
    for j in range(mid+1, high+1):
        s += A[j]
        if(s > right_sum):
            right_sum = s
            max_right = j
    return (max_left, max_right, left_sum+right_sum)

def find_max_subarray(A,low,high):
    if high==low:
        return(low,high,A[low])
    else:
        mid = int((low+high)/2)
        (left_low, left_high, left_sum) = find_max_subarray(A,low,mid)
        (right_low, right_high, right_sum) = find_max_subarray(A,mid+1,high)
        (cross_low, cross_high, cross_sum) = find_max_crossing_subarray(A,low,mid,high)
        if((left_sum >= right_sum) and (left_sum > cross_sum)):
            return(left_low,left_high,left_sum)
        elif((right_sum >= left_sum) and (right_sum > cross_sum)):
            return(right_low,right_high,right_sum)
        else:
            return(cross_low,cross_high,cross_sum)
```

Here we print the low, high and the sum of the maximum subarray to see the function working

```python
A = np.array([1,3,4,1,-4,2,5,6,-3,-6])
print(find_max_subarray(A,0,A.shape[0]-1))
```

```python
n = np.array([10,100,1000,10000,100000])
delta_t = np.array([])
nn = np.linspace(10,100000,100)
y = nn*np.log(nn)/500000

for num in n:
    A = np.random.randint(-10,10,size=num)
    t0 = time.perf_counter()
    subarray = find_max_subarray(A,0,A.shape[0]-1)
    t1 = time.perf_counter()
    delta_t = np.append(delta_t,(t1-t0))
    print(f"n = {num} t = {delta_t[-1]}")
    print("subarray:",subarray)

plt.figure(figsize=(20,10))
plt.subplot(131)
plt.plot(n, delta_t,)
plt.title("Find Maximum Subarray")

plt.subplot(132)
plt.plot(nn, y)
plt.title("nlog(n)")

plt.subplot(133)
plt.plot(nn, y,label="nlog(n)")
plt.plot(n, delta_t,label="Find Maximum Subarray")
plt.title("Both")
plt.legend()
plt.show()
```

The graph on the left shows the time spent for finding the maximum array, the graph on the middle is a graph for C*nlog(n) where C is 1/500000 for scaling, the third graph shows plots both of them in one graph. By looking at the left and the middle graph, we can see that the shape of the graphs are close to being identical. By combining the plots(right graph) we can see that the difference created as n increases between the two plots seem to increase linearly which shows that nlog(n) is the correct theoretical complexity because if it weren't, the increase in the difference between the two plots would not look linear.
