```python
#!/usr/bin/python3
import numpy as np

class Node:
    def __init__(self,key):
        self.left = None
        self.right = None
        self.p = None
        self.key = key
        self.color = "RED"

    def __repr__(self):
        return f"{self.key}"

class RedBlackTree:
    def __init__(self):
        self.nil = Node(None)
        self.nil.color = "BLACK"
        self.root = self.nil
        self.size = 0
    
    def left_rotate(self, x):
        y = x.right
        x.right = y.left
        if y.left != self.nil:
            y.left.p = x
        y.p = x.p
        if x.p == self.nil:
            self.root = y
        elif x == x.p.left:
            x.p.left = y
        else:
            x.p.right = y
        y.left = x
        x.p = y
        
    def right_rotate(self, x):
        y = x.left
        x.left = y.right
        if y.right != self.nil:
            y.right.p = x
        y.p = x.p
        if x.p == self.nil:
            self.root = y
        elif x == x.p.right:
            x.p.right = y
        else:
            x.p.left = y
        y.right = x
        x.p = y

    def insert(self,z):
        y = self.nil
        x = self.root
        while x != self.nil:
            y = x
            if z.key < x.key:
                x = x.left
            else:
                x = x.right
        z.p = y
        if y is self.nil:
            self.root = z
        elif z.key < y.key:
                y.left = z
        else:
            y.right = z
        z.left = self.nil
        z.right = self.nil
        z.color = "RED"
        self.insert_fixup(z)
        self.size += 1

    def insert_fixup(self,z):
        while z.p.color == "RED" and (z.p != self.nil):
            if z.p == z.p.p.left:
                y = z.p.p.right
                if y.color == "RED":
                    z.p.color = "BLACK"
                    y.color = "BLACK"
                    z.p.p.color = "RED"
                    z = z.p.p
                else:
                    if z == z.p.right:
                        z = z.p
                        self.left_rotate(z)
                    z.p.color = "BLACK"
                    z.p.p.color = "RED"
                    self.right_rotate(z.p.p)
            else:
                y = z.p.p.left
                if y.color == "RED":
                    z.p.color = "BLACK"
                    y.color = "BLACK"
                    z.p.p.color = "RED"
                    z = z.p.p
                else:
                    if z == z.p.left:
                        z = z.p
                        self.right_rotate(z)
                    z.p.color = "BLACK"
                    z.p.p.color = "RED"
                    self.left_rotate(z.p.p)
        self.root.color = "BLACK"

    def traverse(self,n):
        l = [(n.key)]
        if n.left.key != None:
            l.append(self.traverse(n.left))
        if n.right.key != None:
            l.append(self.traverse(n.right))
        if n.key != None:
            return l


    def __repr__(self):
        tree_list = self.traverse(self.root)
        return str(tree_list)
        
```

```python
rbt = RedBlackTree()

A = np.random.randint(10,size=5)

for n in A:
    rbt.insert(Node(n))

def printTree(node, level=0):
    if node.key != None:
        printTree(node.left, level + 1)
        print(' ' * 4 * level + '->', node.key)
        printTree(node.right, level + 1)

printTree(rbt.root)
```

```python
rbt = RedBlackTree()

A = np.random.randint(10,size=50)

for n in A:
    rbt.insert(Node(n))

printTree(rbt.root)
```

The above is a representation of a rbtree which grows horizontally, It could be seen that with 50 nodes, the tree is not significantly deeper than the tree with 5 nodes. This is because the rbt algorithm tries to balance the tree at all times. Since searching for a specific node takes O(height of tree) time, balancing the tree really helps keep the operation times low.
