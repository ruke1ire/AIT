```python
from sklearn.datasets import load_iris
from seaborn import pairplot
import matplotlib.pyplot as plt

data = load_iris()
```

```python
print(dir(data))
features = data.feature_names
target_names = data.target_names
X = data.data
y = data.target
```

```python
import pandas as pd

df = pd.DataFrame(X,columns=features)
label_df = pd.Series(y)
df['labels'] = label_df

pairplot(df,hue="labels",markers=["o", "s", "D"])
```

```python
import numpy as np
from IPython.display import clear_output
import time
```

```python
distortion = []
nearests = []
k_list = []
mean_list = []

for k in range(2,10):
    k_list.append(k)
    distortion.append(float('inf'))
    nearests.append(0)
    mean_list.append(0)
    for tries in range(10):
        centers = np.random.uniform(2,4,size=(k,X.shape[1]))
        while True:
            #find the nearest centers for each of the points
            distance = np.empty((X.shape[0],centers.shape[0]))
            for i,x in enumerate(X):
                for j,c in enumerate(centers):
                    distance[i,j] = (c-x).T@(c-x) 
            nearest = np.argmin(distance,axis=1)

            #find the mean of each centers
            mean = centers.copy()
            for i in np.unique(nearest):
                mean[i] = np.mean(X[nearest==i],axis=0)

            if(np.sum((mean-centers)**2)<1e-6):
                dis = np.sum(np.min(distance,axis=1))
                if(distortion[-1]>dis):
                    distortion[-1] = dis
                    nearests[-1] = nearest
                    mean_list[-1] = mean
                break
            else:
                centers = mean

plt.figure(figsize=(10,5))
plt.plot(k_list,distortion)
```

We could see that the elbow is at 3.

```python
df = pd.DataFrame(X,columns=features)
label_df = pd.Series(nearests[k_list.index(3)])
df['labels'] = label_df

pairplot(df,hue="labels",markers=["o", "s", "D"])
```
