```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Here we create a normalized input matrix

```python
# Load student admissions data. The data file does not contain headers,
# so we use hard coded indices for exam 1, exam2, and the admission decision.
data = np.loadtxt('ex2data1.txt',delimiter = ',')
exam1_data = data[:,0]
exam2_data = data[:,1]
X = np.array([exam1_data, exam2_data]).T
y = data[:,2]

mean = np.mean(X,axis=0)
std = np.std(X,axis=0)
X_norm = (X-mean)/(std)

# Output some sample data 
print('Exam scores', X[0:5,:])
print('Normalized',X_norm[0:5,:])
print('-----------------------------')
print('Admission decision', y[0:5])
```

```python
# Plot the data
idx_0 = np.where(y == 0)
idx_1 = np.where(y == 1)

fig1 = plt.figure(figsize=(5, 5)) 
ax = plt.axes()
ax.set_aspect(aspect = 'equal', adjustable = 'box')
plt.title('Distribution')
plt.xlabel('Exam 1')
plt.ylabel('Exam 2')
plt.grid(axis='both', alpha=.25)
ax.scatter(exam1_data[idx_0], exam2_data[idx_0], s=50, c='r', marker='*', label='Not Admitted')
ax.scatter(exam1_data[idx_1], exam2_data[idx_1], s=50, c='b', marker='o', label='Admitted')
plt.show()
```

```python
import random

# Partion data into training and test datasets

m, n = X.shape
XX = np.insert(X, 0, 1, axis=1)
XX_norm =  np.insert(X_norm,0,1,axis=1)

y = y.reshape(m, 1)
idx = np.arange(0, m)
random.shuffle(idx)
percent_train = .6
m_train = int(m * percent_train)
train_idx = idx[0:m_train]
test_idx = idx[m_train:]

X_train = XX[train_idx,:];
X_test = XX[test_idx,:];
X_train_norm = XX_norm[train_idx,:]
X_test_norm = XX_norm[test_idx,:]

y_train = y[train_idx];
y_test = y[test_idx];
```

```python
def sigmoid(z):   
    return 1 / (1 + np.exp(-z))

def h(X, theta):
    return sigmoid(X @ theta)

def grad_j(X, y, y_pred):
    return X.T @ (y - y_pred) / X.shape[0]
    
def j(theta, X, y):    
    y_pred = h(X, theta)
    error = (-y * np.log(y_pred+0.00001)) - ((1 - y) * np.log(1.00001 - y_pred))
    cost = sum(error) / X.shape[0]
    grad = grad_j(X, y, y_pred)
    return cost[0], grad
```

```python
theta_initial = np.zeros((n+1, 1))
theta_initial_2 = np.array([-10,0,0]).reshape(-1,1)
print(theta_initial)
print(theta_initial_2)
```

```python
def train(X, y, theta_initial, alpha, num_iters):
    theta = theta_initial
    j_history = []
    for i in range(num_iters):
        cost, grad = j(theta, X, y)
        theta = theta + alpha * grad
        j_history.append(cost)
    return theta, j_history
```

```python
alpha = .0005
num_iters = 1000000
theta, j_history = train(X_train, y_train, theta_initial, alpha, num_iters)

print("Theta optimized:", theta)
print("Cost with optimized theta:", j_history[-1])

plt.plot(j_history)
plt.xlabel("Iteration")
plt.ylabel("$J(theta)$")
plt.title("Training cost over time with batch gradient descent (no normalization)")

plt.show()
```

```python
def r_squared(y, y_pred):
    return 1 - np.square(y - y_pred).sum() / np.square(y - y.mean()).sum()

y_test_pred_soft = h(X_test, theta)
y_test_pred_hard = (y_test_pred_soft > 0.5).astype(int)

test_rsq_soft = r_squared(y_test, y_test_pred_soft)
test_rsq_hard = r_squared(y_test, y_test_pred_hard)
test_acc = (y_test_pred_hard == y_test).astype(int).sum() / y_test.shape[0]

print('Got test set soft R^2 %0.4f, hard R^2 %0.4f, accuracy %0.2f' % (test_rsq_soft, test_rsq_hard, test_acc))
```

The test above is for [0,0,0] initial theta, without normalization. It could be seen that we took one million iterations to get a cost of approximately 0.2305. In the next test, we will pick an initial theta which is already close to the optimal theta, and observe what happens.

```python
alpha = .0005
num_iters = 50
theta_2, j_history_2 = train(X_train, y_train, theta_initial_2, alpha, num_iters)

print("Theta optimized:", theta_2)
print("Cost with optimized theta:", j_history_2[-1])

plt.plot(j_history_2)
plt.xlabel("Iteration")
plt.ylabel("$J(theta)$")
plt.title("Training cost over time with batch gradient descent (no normalization starting point close to optimum)")

plt.show()
```

```python
y_test_pred_soft = h(X_test, theta_2)
y_test_pred_hard = (y_test_pred_soft > 0.5).astype(int)

test_rsq_soft = r_squared(y_test, y_test_pred_soft)
test_rsq_hard = r_squared(y_test, y_test_pred_hard)
test_acc = (y_test_pred_hard == y_test).astype(int).sum() / y_test.shape[0]

print('Got test set soft R^2 %0.4f, hard R^2 %0.4f, accuracy %0.2f' % (test_rsq_soft, test_rsq_hard, test_acc))
```

Interestingly, just changing the initial theta to [-10,0,0] reduced the number of iterations from 1000000 to 50 without a cost of 0.2513, which is comparable to the first test. The next test will see whether normalizing the data improves the learning in terms of cost and number of iterations.

```python
alpha = 10
num_iters = 50
theta_norm, j_history_norm = train(X_train_norm, y_train, theta_initial, alpha, num_iters)

print("Theta optimized:", theta_norm)
print("Cost with optimized theta:", j_history_norm[-1])

plt.plot(j_history_norm)
plt.xlabel("Iteration")
plt.ylabel("$J(theta)$")
plt.title("Training cost over time with batch gradient descent (normalized)")

plt.show()
```

```python
y_test_pred_soft = h(X_test_norm, theta_norm)
y_test_pred_hard = (y_test_pred_soft > 0.5).astype(int)

test_rsq_soft = r_squared(y_test, y_test_pred_soft)
test_rsq_hard = r_squared(y_test, y_test_pred_hard)
test_acc = (y_test_pred_hard == y_test).astype(int).sum() / y_test.shape[0]

print('Got test set soft R^2 %0.4f, hard R^2 %0.4f, accuracy %0.2f' % (test_rsq_soft, test_rsq_hard, test_acc))
```

The normalized test resulted in having a cost of 0.1643 which is significantly lower than both the tests without normalization. Furthermore, the number of iterations remained very low(50).
