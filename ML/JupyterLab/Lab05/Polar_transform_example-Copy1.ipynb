### Polar transformation example

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
### Generate 2 independent nonlinear datasets 

n = 200;

t = np.random.uniform(0,10,n);

# generate class 1 as a function of t
x1 = .25 * np.power(t,3) - .15 * np.power(t,2)  + (np.random.randn(n) *50) + 50;

# generate class 2 as a function of t
x2 = -.25 * np.power(t,3) - .15 * np.power(t,2) + (np.random.randn(n) *50) - 50;
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x1,x2,t,edgecolors='black',c=t,alpha=1)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('t')
ax.view_init(elev=0, azim=90)
```

Each cartesian coordinate can be represented as a polar coordinate using following mathematical relation


$ x = rcos\theta$,  $y = rsin\theta$


where $r$ is the shortest distance from origin, and $\theta$ is the angle from $x$ axis rotated anticlockwise.

```python
# Perform polar transform

r = np.sqrt((x1)**2 + (t)**2 + (x2)**2)

angle1 = np.arctan2(t,x1)
angle2 = np.arctan2(t,x2)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(angle1,angle2,r,edgecolors='black',c=t,alpha=1)
ax.set_xlabel('angle1')
ax.set_ylabel('angle2')
ax.set_zlabel('r')
ax.view_init(elev=0, azim=90)

plt.title('3D plot after polar transform')
plt.show()
```
